```python
import glob
```

```python
list_of_steps = glob.glob(os.path.join(os.getcwd(),"..","**","*.step"), recursive=True)
```

```python
# for step in glob:
#     print(os.path.abspath(step))

list_of_steps = [os.path.abspath(step) for step in list_of_steps]
```

```python
from py_sas_studio_custom_steps import CustomStep
for step in list_of_steps:
    step_directory = os.path.dirname(step)
    file_name = os.path.basename(step)
    sas_file_name = os.path.join(os.path.dirname(step),"extras",os.path.basename(os.path.splitext(step)[0]+".sas"))
    print(file_name)
    print(sas_file_name)
    custom_step = CustomStep()
    step_program = custom_step.extract_sas_program(step)
    if not os.path.exists(os.path.join(step_directory,"extras")):
        os.makedirs(os.path.join(step_directory,"extras"))
        print(f"Extras directory created in {step_directory}")
    else:
        print(f"Extras directory exists in {step_directory}")
    if not os.path.exists(sas_file_name):
        with open(sas_file_name,"w") as f:
            f.write(step_program)
        print(f"SAS program created in {step_directory}")

    
```
